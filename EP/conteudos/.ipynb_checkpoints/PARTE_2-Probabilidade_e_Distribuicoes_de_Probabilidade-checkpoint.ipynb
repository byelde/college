## Introdução 

* Na aula anterior aprendemos sobre a diferença entre entre população e
amostra
  * população - todas as observações possíveis de um determinado fenômeno (universo)
  * amostra - subconjunto da população
* Quando temos condições de observar toda uma população, utilizamos técnicas da estatı́stica descritiva para entender o fenômeno
* Quando não conseguimos observar toda a população, utilizamos técnicas de amostragem para selecionar uma amostra daquela população
* A partir da amostra colhida, podemos tirar conclusões dessa amostra e fazer inferência para toda a população
  * analisamos a amostra e tiramos conclusões
  * generalizamos para a população, acreditando que o que se observa para
a amostra é verdadeiro para toda a população
* A validade das conclusões depende da qualidade de amostragem usada para selecionar os dados da amostra


## Noções de Amostragem
* Utilizando téncicas de amostragem apropriadas, evitamos produzir resultados tendenciosos
  * Um estudo realizado com dados falhos resulta em resultados questionáveis
* A suposição básica de um estudo amostral válido é que a amostra representa a população
* O objetivo é minimizar o erro de amostragem, que é a diferença entre
os resultados da amostra e da população
* Principais técnicas de amostragem
  * Amostragem aleatória
  * Amostragem estratificada
  * Amostragem sistemática

### Amostragem Aleatória
* Todos os membros de uma população têm chances iguais de serem selecionados
* Tem grandes chances de selecionar um grupo amostral que seja representativo da população de estudo
* Para se selecionar uma amostra de tamanho n de uma população de tamanho N, basta selecionar aleatoriamente os n elementos sem colocá-los de volta no sorteio
* Ex: Deseja-se avaliar a qualidade do ensino em um curso de graduação à distância
  * O curso dispõe de 500 alunos, mas é possı́vel avaliar apenas 50 destes alunos
  * Utilizando a técnica de amostragem, podemos fazer um sorteio onde cada um dos 500 alunos possui mesma probabilidade de ser avaliado

```
#criano uma lista com o número dos 500 alunos
alunos = [i for i in range(500)]
alunos
```

```
import random

#selecionando 50 alunos sem repetição
alunos_selecionados = random.sample(alunos,50)
alunos_selecionados
```

### Amostragem Estratificada
* Essa técnica de amostragem leva em consideração as caracterı́sticas
da população
* Digamos que uma população tem vários segmentos e a amostra deve
ser representativa de todos estes segmentos
  * utilizar a amostragem aleatória pode resultar em dados que não são
representativos
* Neste caso é mais indicado o uso da amostragem estratificada
* Ex: pesquisa de intenções de voto
  * Em um processo eleitoral, as diferentes classes de eleitores têm
intenções de voto influenciadas por diversas caracterı́sticas
    * idade, sexo, condição financeira, nı́vel de instrução, etc
  * Uma pesquisa feita somente com pessoas de mesma classe social e
nı́vel de instrução pode ser tendenciosa, pois não reflete toda a
população
  * Neste caso, uma amostragem estratificada seleciona pessoas com
caracterı́sticas diferentes para responder a pesquisa
    * garante-se uma amostra mais próxima do que seria a população
* Método de realização de amostragem estratificada:
  * Divide-se a população em grupos com caracterı́sticas similares
  * Escolhe-se aleatoriamente elementos de cada um dos estratos
* A amostra estratificada assegura que cada segmento da população
será representado
* Em seguida, selecionarı́amos aleatoriamente elementos em cada um
dos grupos
* Terı́amos representantes aleatórios de todas as classes ou extratos da
população

```
#criando listas dos alunos de cada sala da escola
sala1 = ['antonio','pedro','ana','joao','guilherme','carlos','maria']
sala2 = ['vitoria','cris','jose','simão','julio']
sala3 = ['lais','mariana','cicero','yuri','jonathan','vitor']
```

```
#selecionando três alunos de cada turma
alunos_selecionados = []
alunos_selecionados.append(random.sample(sala1,3))
alunos_selecionados.append(random.sample(sala2,3))
alunos_selecionados.append(random.sample(sala3,3))
alunos_selecionados
```

### Amostra sistemática
* Uma amostra sistemática é aquela na qual é atribuı́do um número a
cada membro da população
* Os membros são então ordenados segundo algum critério de
ordenação
* Um número inicial é selecionado aleatoriamente e membros da
amostra são selecionados em intervalos regulares a partir do número
inicial
* Ex: seja uma contagem do número de pessoas por residência
  * poderı́amos designar um número diferente para cada residência
  * escolher aleatoriamente um número inicial
  * selecionar cada décima residência a partir da primeira e contar o número de pessoas vivendo
em cada uma

```
residencias = [i for i in range(100)]
residencias
```

```
residencias_selecionadas = [residencias[i] for i in range(0,100,10)]
residencias_selecionadas
```

## Variáveis Aleatórias
* Podemos ver que em todos os casos de amostragem temos um fator
probabilı́stico envolvido
* Selecionar um item aleatório de um conjunto envolve a probabilidade
deste item ser escolhido
* O resultado de um experimento de probabilidade geralmente é uma
contagem ou medida
* Quando isso ocorre, o resultado é chamado de variável aleatória
* Uma variável aleatória $x$ representa um valor numérico associado a
cada resultado de um experimento de probabilidade
* A palavra "aleatória" indica que $x$ é determinado pelo acaso
* A média de uma amostra escolhida ao acaso de uma população é uma
variável aleatória
  * seus valores variam de amostra para amostra
* Ex: Um sistema de comunicação por voz tem 48 linhas externas. Em um dado instante, entre 0 e 48 linhas podem estar ocupadas. 
  * neste caso o número de linhas ocupadas x é uma variável aleatória
  * x pode assumir valores inteiros entre 0 e 48
* Ex: Seja x o valor em um dado não viciado.
  * valores possı́veis para x são \{1, 2, 3, 4, 5, 6\}
  * a probabilidade de qualquer valor de x é igual a 1/6
* Ex: Sejam as idades das pessoas de uma famı́lia de 20 integrantes
  * 12 15 25 19 45 52 63 72 5 37 7 12 31 40 2 61 69 33 25 9
  * A média populacional desse conjunto é dado pela soma dos valores
dividida pelo número de valores 

    $\mu = \dfrac{\sum x}{N} = \dfrac{634}{20} = 31$
  * Esse valor para a média será o mesmo sempre e não há incerteza sobre isso
  * No entanto, se sortearmos aleatoriamente 5 pessoas e calcularmos a média de idade da amostra, o valor da média amostral pode mudar, dependendo da amostra
    * seja a amostra 15 31 63 69 37, temos
    
      $\overline{x} = \dfrac{15 + 31 + 63 + 69 + 37}{5} = \dfrac{215}{5} = 43$

    * seja a amostra 12 15 5 2 25

      $\overline{x} = \dfrac{12 + 15 + 5 + 2 + 25}{5} = \dfrac{59}{5} = 11.8$

    * como a média amostral varia de amostra para amostra, a média nessa caso é uma variável aleatória

* Há dois tipos de variáveis aleatórias: variáveis discretas e contı́nuas
* Uma variável aleatória discreta é aquela que tem um número finito ou
contável de possı́veis resultados a serem listados
* Uma variável aleatória contı́nua é aquela que tem um número
incontável de possı́veis resultados, representados por um intervalo na
reta numérica
* É importante saber diferenciar entre os dois tipos, pois utilizaremos técnicas estatísticas diferentes para variáveis discretas ou contínuas
    

## Distribuições de Probabilidade Discretas

* Para cada valor de uma variável aleatória discreta pode-se determinar a probabilidade para a ocorrencia desse valor
* Ao listar cada valor de uma variável aleatória com sua probabilidade
correspondente, você estará formando uma distribuição de probabilidade
* Uma distribuição de probabilidade discreta lista cada valor possı́vel
que a variável aleatória pode assumir, junto com sua probabilidade
* Condições de uma distribuição de probabilidade
  * A probabilidade de cada valor da variável aleatória discreta está entre 0
e 1
  * A soma de todas as probabilidades é 1

## Distribuição de Bernoulli
* Existem muitos experimentos que adimitem apenas dois resultados:
sucesso ou fracasso
  * ex: jogador de basquete irá fazer um lance livre. Possibilidades:
acertar ou não
  * ex: jogar uma moeda para cima. Possibilidades: dar cara ou não
(coroa)
  * ex: o diagnóstico de um exame médico para detecção de uma doença.
Possibilidades: positivo ou negativo
* Experimentos de probabilidade como esses são chamados de ensaios
de Bernoulli
  * variável aleatória admite apenas dois valores
  * esses valores estão associados a probabilidades de ocorrer
* Em um experimento de Bernoulli há apenas dois resultados possı́veis:
sucesso (p) e fracasso (q)
* A probabilidade de um sucesso P(S) é a mesma em cada tentativa
* Como só existem duas possibilidades de resultado, a soma da
probabilidade de sucesso e de fracasso é igual a 1
  * p + q =1
* Dessa forma:
  * q =1 − p
* A variável aleatória $x$ contabiliza o número de tentativas com sucesso
* É possível demonstrar que para uma variável aleatória de Bernoulli
  * a média para essa distribuição de probabilidade é dada por:
    * $\overline{x} = E(x) = p$
  * a variâncai para essa distribuição de probabilidade é dada por:
    * $s^2 = VAR(x) = p(1-p) = p\cdot q$
* Ex: Uma urna tem 30 bolas brancas e 20 verdes. Retira-se uma bola dessa
urna. Seja X o número de bolas verdes. Calcule $E(X)$ e $VAR(X)$
  * $E(X) = p = \dfrac{20}{50} = \dfrac{2}{5} = 0.4$
  * $VAR(X) = p\cdot q = \dfrac{20}{50} \cdot \dfrac{30}{50} = \dfrac{60}{250} = 0.24$

```
verdes = [1]*20
```

## Distribuição Binomial
* Uma outra importante distribuição de probabilidade discreta é a
Binomial
  * estudo de ensaios de Bernoulli repetidas vezes
* Ex: Suponha que uma moeda é lançada 3 vezes e a probilidade de
cara seja $p$ em cada lançamento. Qual a distribuição de probabilidade
do número de caras em 3 lançamentos?
  * variável $x$ é o número de caras - distribuição binomial
  * $x_i$ é uma variável aleatória de Bernoulli ($i = \{1, 2, 3\}$)
  * Possibilidades: \{FFF , FFS, FSF , SFF , FSS, SFS, SSF , SSS\}
* Podemos calcular a probabilidade de cada evento utilizando uma árvore de possibilidades

image.png

* Em um experimento binomial, a probabilidade de exatamente $x$ sucessos em $n$ tentativas é dado por:

  $P(x) = \dfrac{n!}{(n-x)!x!}p^xq^{n-x}$
* Ex: Cirurgias de microfraturas no joelho têm 75% (3/4) de chance de
sucesso com pacientes com joelhos degenerativos
  * A cirurgia é realizada em três pacientes
  * Qual a probabilidade de a cirurgia ser um sucesso em exatamente dois
pacientes?
  * A solução para esse experimento pode ser vista como uma árvore de
possibilidades:

image.png


* Há três resultados em que ocorrem dois sucessos e cada um tem probabilidade $\dfrac{9}{64}$ de ocorrer
* Logo a probabilidade de sucesso em exatamente dois pacientes é de $3\cdot \dfrac{9}{64} = 0.422$ ou $42.2\%$
* Utilizando a fórmula para calcular a probabilidade do evento temos:

  $P(2) = \dfrac{n!}{(n-x)!x!}\cdot p^xq^{n-x} = \dfrac{3!}{(3-2)!2!}\Big(\dfrac{3}{4}\Big)^2\Big(\dfrac{1}{4}\Big)^1$

* Para se construir uma distribuição de probabilidade binomial, é preciso
encontrar os valores possı́veis para a variável aleatória x e em seguida
encontrar a probabilidade correspondente a cada um destes valores
* Ex: Em uma pesquisa com trabalhadores de uma determinada
empresa, lhes foi perguntado qual a fonte de rendimentos esperada
por estes durante a aposentadoria
* Sabe-se que 25% dos trabalhadores da empresa esperam contar com a
previdência social como fonte de seus rendimentos
* Sete trabalhadores foram escolhidos aleatoriamente e lhes foi
perguntado se esperam contar com a previdência social durante a
aposentadoria
* Neste caso, o sucesso é o trabalhador responder sim e o fracasso é o
trabalhador responder não
  * p = 0.25 e q = 0.75

```
from math import factorial
```

```
def probabilidadeBinomial(x,n,p,q):
  return (factorial(n)/(factorial(n-x)*factorial(x)))*(p**x)*(q**(n-x))
```

```
p = 0.25
q = 0.75
n = 7
probabilidades = []
for x in range(8):
  px = probabilidadeBinomial(x,n,p,q)
  print('P(',x,')=',px)
  probabilidades.append(px)
```

```
np.sum(probabilidades)
```

```
#histograma de frequência
x = [0,1,2,3,4,5,6,7]
plt.bar(x,probabilidades)
plt.xlabel('x')
plt.ylabel('P(x)')
plt.title('Distribuição de Probabilidade')
plt.grid(True)
plt.show()
```

* Para qualquer distribuição binomial, podemos encontrar a média, a
variância e o desvio padrão desta distribuição, ou seja, seus
parâmetros amostrais
* A média é dada por
  
  $\overline{x} = E(x) = n \cdot p$
* A variância é dada por
  
  $s^2 = VAR(x) = n \cdot p \cdot q$

* Ex: em uma determinada cidade brasileira, cerca de 56% dos dias são
nublados. Vamos encontrar a média a variância e o desvio padrão
para o número de dias nublados durante o mês de junho
  * O mês de junho tem 30 dias, logo n = 30
  * p = 0.56 e q = 1 - p = 1 - 0.56 = 0.44
  * Aplicando as fórmulas podemos encontrar os parâmetros:  
  * $\overline{x} = E(X) = n \cdot p = 30 \cdot 0.56 = 16.8$
  * $s^2 = VAR(x) = n \cdot p \cdot q = 30 \cdot 0.56\cdot 0.44 = 7.4$
  * $s = \sqrt{s^2} = \sqrt{7.4} = 2.7$
