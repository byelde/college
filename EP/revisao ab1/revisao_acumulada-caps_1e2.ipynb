Em 2012, um estudo mundial com todas as companhias aéreas descobriu que os atrasos nas bagagens eram causados por manuseio incorreto na chegada (4%), falha ao carregar (15%), erro de carregamento (5%), restrição de tamanho e peso (7%), erro de identificação (3%), manuseio incorreto na transferência (53%) e erro debilhete / troca / segurança / outros (13%). Use um gráfico de Pareto para organizar os dados. (Fonte: SocietyInternational de Telecomunications Aeronautics.)

```python
data: dict = {
  "causa":["manuseio incorreto na chegada", "falha ao carregar", "erro de carregamento", "restrição de tamanho e peso", "erro de identificação", "manuseio incorreto na transferência", "erro debilhete/troca/segurança/outros"],
  "frequencia":[4, 15, 5, 7, 3, 53, 13]
}

df = pd.DataFrame(data)
df.sort_values(by="frequencia", inplace=True, ascending=False)
```

```python
df.plot(kind='bar', x='causa', y='frequencia')
```
